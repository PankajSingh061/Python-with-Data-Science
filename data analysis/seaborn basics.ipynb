```python
import numpy as np
import pandas as pd
import seaborn as sns
from matplotlib import pyplot as plt
```

```python
tips=sns.load_dataset('tips')
```

```python
tips.head(5)
```

```python
tips.shape
```

```python
sns.set_theme(style='darkgrid')
```

```python
sns.relplot(data=tips,x='total_bill',y='tip')
```

```python
#adding of a third variable is called hue
sns.relplot(data=tips,x='total_bill',y='tip',hue='smoker')
```

```python
sns.relplot(data=tips,x='total_bill',y='tip',hue='day')
```

```python
sns.relplot(data=tips,x='total_bill',y='tip',hue='sex')
```

```python
sns.relplot(data=tips,x='total_bill',y='tip',hue='time')
```

```python
tips.head()
```

```python
# if the data values is numeric then the hue switches its colour plattes
sns.relplot(data=tips,x='total_bill',y='tip',hue='size')
```

```python
#To emphasize the difference between the classes, and to improve accessibility, you can use a different marker style for each class:

sns.relplot(data=tips,x='total_bill',y='tip',hue='smoker',style='smoker')
```

```python
sns.relplot(data=tips,x='total_bill',y='tip',hue='smoker',style='time')
```

```python
sns.relplot(data=tips,x='total_bill',y='tip',hue='size',palette='rocket')
```

```python
#change size of each point
sns.relplot(data=tips,x='total_bill',y='tip',size='size')
```

```python
sns.relplot(data=tips,x='total_bill',y='tip',size='size',sizes=(15,200))
```
